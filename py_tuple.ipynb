Create a nested tuple and iterate over the elements, printing each element.

```python
def print_nested_tuple(tup):
    for element in tup:
        if isinstance(element, tuple):
            print_nested_tuple(element)
        else:
            print(element)

nt = (1,(2,(3,(4,(5,(6))))))
print_nested_tuple(nt)
```
